```python
# Making the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)

# Preprocessing Input data
data = pd.read_csv('data.csv')
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
plt.scatter(X, Y)
plt.show()

x = X.tolist()
y = Y.tolist()
```

```python
def linearRegressionModel(stepSize = 0.001, maxIteration=1000, x = [], y=[], n = None):
    c=0 #c
    m=0 #m
    for _ in range(maxIteration):
        sumDiffrenceC = 0
        sumDiffrenceM = 0
        for i in range(n):
            sumDiffrenceC += (c + (m* x[i])) - y[i]
            sumDiffrenceM += ((c + (m* x[i])) - y[i])*x[i]
        c = c-(stepSize * (1/n) * sumDiffrenceC)
        m = m-(stepSize * (1/n) * sumDiffrenceM)
    return c, m
c, m = linearRegressionModel(x=x, y=y, n=len(x))
print(c, m)
```

```python
# Making predictions
y_pred = []
for i in range(len(y)):
    y_pred.append(m*x[i] + c)

plt.scatter(x, y)
plt.scatter(x, y_pred, color='red')
#plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') # predicted
plt.show()
```

m is the number of samples in the training set
